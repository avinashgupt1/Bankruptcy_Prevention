```python
df.columns = ["Indstrial_Risk","Management_Risk","Financial_Flexibility","Credibility","Competitive","Operating_Risk","Class"]
from sklearn import preprocessing
df["Class"] = preprocessing.LabelEncoder().fit_transform(df["Class"])
```

```python
df
```

```python
import seaborn as sns
from sklearn.decomposition import PCA
from sklearn.preprocessing import scale
```

```python
pca = PCA(3)
pca_values = pca.fit_transform(df.drop(["Class"],1))
```

```python
pca_values
```

```python
pca_values[:,0]
```

```python
final_Df = pd.DataFrame()
final_Df["PCA1"]= pca_values[:,0]
final_Df["PCA2"] = pca_values[:,1]
final_Df["PCA3"] = pca_values[:,2]
final_Df["Class"] = df["Class"]
```

```python
final_Df
```

```python
import matplotlib.pyplot as plt
var = pca.explained_variance_ratio_
var
plt.bar(range(1,len(var)+1),var)
```

```python
var
```

```python
from sklearn.ensemble import RandomForestClassifier
X = final_Df.drop(["Class"],1)
y = final_Df["Class"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,y,train_size=0.8,shuffle=True)
```

```python
from sklearn.metrics import accuracy_score
rf = RandomForestClassifier()
rf.fit(X_train,Y_train)
pred_train = rf.predict(X_train)
accuracy_score(Y_train,pred_train)
```

```python
pred_test = rf.predict(X_test)
accuracy_score(Y_test,pred_test)
```
